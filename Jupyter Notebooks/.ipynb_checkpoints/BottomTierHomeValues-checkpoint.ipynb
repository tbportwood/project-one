```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
bottom_tier = pd.read_csv("BottomTierHomeValues.csv")
bottom_tier.head()
```

```python
fig = plt.figure()
fig.set_size_inches(11,8)

ax = fig.add_subplot() 
ax2 = ax.twinx() 

width = 0.4
x_axis = bottom_tier["RegionName"]

bottom_tier["InitialValue"].plot(kind='bar', color='red', ax=ax, width=width, position=1)
bottom_tier["FinalValue"].plot(kind='bar', color='blue', ax=ax, width=width, position=0)
ax.set_xticklabels(x_axis)

ax.set_ylabel('Initial Value')
ax2.set_ylabel('Final Value')
ax.set_ylim(0, 400000)
ax2.set_ylim(0, 400000)

plt.savefig("Bottom Tier Home Values")
```

```python
fig = plt.figure()
fig.set_size_inches(11,8)

ax = fig.add_subplot() 

width = 0.4
x_axis = bottom_tier["RegionName"]

bottom_tier["Percent Change"].plot(kind='bar', color='blue', ax=ax, width=width, position=0)
ax.set_xticklabels(x_axis)

ax.set_ylabel('Percent Change')

plt.savefig("Bottom Tier Percent Change")
```
