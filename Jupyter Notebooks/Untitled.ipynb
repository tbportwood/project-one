```python
import pandas as pd
import requests
import json
from api_keys import g_key
```

```python
neighborhoods = pd.read_csv("HomeValues.csv")
neighborhoods
```

```python
neighborhoods["Lat"] = ""
neighborhoods["Lng"] = ""
neighborhoods
```

```python
params = {"key": g_key}


for index, row in neighborhoods.iterrows():
    base_url = "https://maps.googleapis.com/maps/api/geocode/json"

    region = row['RegionName']
    state = row['State']


    params['address'] = f"{region},{state}"

   
    neighborhoods_lat_lng = requests.get(base_url, params=params)
    

    print(neighborhoods_lat_lng.url)
    

    neighborhoods_lat_lng = neighborhoods_lat_lng.json()

    neighborhoods.loc[index, "Lat"] = neighborhoods_lat_lng["results"][0]["geometry"]["location"]["lat"]
    neighborhoods.loc[index, "Lng"] = neighborhoods_lat_lng["results"][0]["geometry"]["location"]["lng"]


neighborhoods.head()
```
